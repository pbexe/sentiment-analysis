```python
import matplotlib.pyplot as plt
import os

#from google.colab import drive
#drive.mount('/content/drive/')
#path = '/content/drive/MyDrive/Uni/Masters/Machine Learning - Natural Language/Portfolio 2/sentiment-analysis-main/' #please use whatever drive/folder the data set is stored in here
#os.chdir(path)

def distribution():
  labels = open('../data/train_labels.txt', 'r')
  training_labels = labels.read().split('\n')
  labels.close()

  distributed_labels = []
  label_0 = 0
  label_1 = 0
  label_2 = 0

  for label in training_labels:
    if label == "0":
      label_0 += 1
    if label == "1":
      label_1 += 1
    if label == "2":
      label_2 += 1

  distributed_labels.append(label_0)
  distributed_labels.append(label_1)
  distributed_labels.append(label_2)

  y_pos = [0,1,2]
  y_labels = ["Negative","Neutral","Positive"]

  fig, ax = plt.subplots()

  hbars = ax.barh(y_pos, distributed_labels, align='center')
  ax.set_yticks(y_pos)
  ax.set_yticklabels(y_labels)
  ax.invert_yaxis
  ax.set_xlabel('Frequency')
  ax.set_title('Sentiment Distribution')

  plt.show()

distribution()

 # lengths = df_np['tweet_length'].plot(
#    kind='bar',
#    title='Sentiment Distribution of Training Set')

  #plt.show()
```
